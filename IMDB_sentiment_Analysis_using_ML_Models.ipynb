#Data Cleaning

```python
# check the null values in data
df.isna().sum()
```

```python
import re

def process(review):
    review = review.lower()
    review = re.sub("[^a-zA-Z]", ' ', review)
    review = re.sub(r"https\S+|www\S+|http\S+", '', review, flags=re.MULTILINE)
    review = re.sub(r'\@\w+|\#', '', review)
    review = re.sub(r'[^\w\s]', '', review)
    return review

# Apply the process function to the 'review' column of your DataFrame
df['review'] = df['review'].apply(process)
```

```python
df.head(10)
```

```python
duplicated_count = df.duplicated().sum()
print("Number of duplicate entries: ", duplicated_count)
```

```python
df = df.drop_duplicates('review')
```

```python
nltk.download('punkt')
```

```python
def tokenize(column):

    tokens = nltk.word_tokenize(column)
    return [w for w in tokens if w.isalpha()]
```

```python
df['tokenized'] = df.apply(lambda x: tokenize(x['review']), axis=1)
```

```python
df.head()
```

```python
nltk.download('stopwords');
```

```python
#Stopwords removal
def remove_stopwords(tokenized_column):
    stops = set(stopwords.words("english"))
    return [word for word in tokenized_column if not word in stops]
```

```python
df['stopwords_removed'] = df.apply(lambda x: remove_stopwords(x['tokenized']), axis=1)
df[['review', 'stopwords_removed']].head()
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
def apply_stemming(tokenized_column):
    stemmer = PorterStemmer()
    return [stemmer.stem(word) for word in tokenized_column]
```

```python
df['Stemming'] = df.apply(lambda x: apply_stemming(x['stopwords_removed']), axis=1)
df[['review', 'Stemming']].head()
```

```python
def rejoin_words(tokenized_column):
    return ( " ".join(tokenized_column))
```

```python
df['Clean_Data'] = df.apply(lambda x: rejoin_words(x['Stemming']), axis=1)
df[['review', 'Clean_Data']].head()
```

```python
df.head()
```

```python
columns_to_drop = ['tokenized', 'stopwords_removed', 'Stemming']
df.drop(columns=columns_to_drop, inplace=True)
```

#Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer()
bow = tv.fit_transform(df['Clean_Data'])
bow = np.array(bow.todense())
```

```python
X = bow
y = df['sentiment']
```

#Train_Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

#Train Machine Learning Models

```python
mnb = MultinomialNB().fit(X_train, y_train)
y_pred = mnb.predict(X_test)
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('F1 score:', f1_score(y_test, y_pred, average="macro"))
print('ROC AUC:', roc_auc_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Random Forest Classifier

rfc = RandomForestClassifier(n_estimators = 10, random_state = 0)
rfc=rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
```

```python
print('Accuracy:', accuracy_score(y_test, rfc_pred))
print('F1 score:', f1_score(y_test, rfc_pred, average="macro"))
print('ROC AUC:', roc_auc_score(y_test, rfc_pred))
```

```python
print(classification_report(y_test, rfc_pred))
```

```python
from sklearn.linear_model import LogisticRegression
 #Logistic Regression
lr = LogisticRegression().fit(X_train, y_train)
lr_pred = lr.predict(X_test)
```

```python
print('Accuracy:', accuracy_score(y_test, lr_pred))
print('F1 score:', f1_score(y_test, lr_pred, average="macro"))
print('ROC AUC:', roc_auc_score(y_test, lr_pred))
```

```python
print(classification_report(y_test, lr_pred))
```

```python
# Using linear support vector classifier
from sklearn.svm import LinearSVC
lsvc = LinearSVC().fit(X_train, y_train)
lsvc_pred = lsvc.predict(X_test)
```

```python
print('Accuracy:', accuracy_score(y_test, lsvc_pred))
print('F1 score:', f1_score(y_test, lsvc_pred, average="macro"))
print('ROC AUC:', roc_auc_score(y_test, lsvc_pred))
```

```python
print(classification_report(y_test, lsvc_pred))
```

```python
from sklearn import  metrics # import library for getting scores

models = []
models.append(('MNB', mnb))
models.append(('LR', lr))
models.append(('RF', rfc))
models.append(('SVC', lsvc))
precision_score = []
recall_score = []
f1_score = []
accuracy_score = []
roc_auc_score = []
for name, model in models:
    print(name)
    print("precision_score: {}".format(metrics.precision_score(y_test , model.predict(X_test), average='weighted') ))
    print("recall_score: {}".format( metrics.recall_score(y_test , model.predict(X_test), average='weighted') ))
    print("f1_score: {}".format( metrics.f1_score(y_test , model.predict(X_test), average='weighted') ))
    print("accuracy_score: {}".format( metrics.accuracy_score(y_test , model.predict(X_test)) ))


    precision_score.append(metrics.precision_score(y_test , model.predict(X_test), average='weighted') )
    recall_score.append(metrics.recall_score(y_test , model.predict(X_test), average='weighted') )
    f1_score.append( metrics.f1_score(y_test , model.predict(X_test), average='weighted') )
    accuracy_score.append(metrics.accuracy_score(y_test , model.predict(X_test)))
```
